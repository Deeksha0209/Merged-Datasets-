```python
#Removing the extra/not required columns
df1=df1.drop(columns =['Indicator Name','Indicator Code'])
print(df1)
```

```python
#Analysed the df & found out that majority countries had more than half data filled hence will substitute NA values in those countries with the row_avg.
#The remaining countries that have NaN values in all/more than half cols will be dropped.
row_avg = df1.mean(axis=1)
print(row_avg)
```

```python
#after analysing df, I found out that countries that have nan val are i.e to be dropped - Andorra[4], American[9],caymen[50],Gibraltor[82],Not classified[108],st.kitts[123],Monaco[147],marshallis[153],northern[162]
#Nauru[177],San Marina[210],Turks[226],Taluvu[243],Brits[253],Kosko[259]
df1.loc[253]
```

```python
#Lets first drop these rows!
df1=df1.drop([df1.index[4],df1.index[9],df1.index[50],df1.index[82],df1.index[108],df1.index[123],df1.index[147],df1.index[153],df1.index[162],df1.index[177],df1.index[210],df1.index[226],df1.index[243],df1.index[253],df1.index[259]])
print(df1)
```

```python
#now I will substitue the rest of the rows(that have more than half colm filled ) i.e having just few NaN values with row_avg using Transpose (T)
df1=df1.T.fillna(df1.mean(axis=1)).T
```

```python
#Now, lets chk the updated isnull values. Should be 0 for all!
df1.isnull().sum()

#Yayy!
```

```python
#Now resetting index so that the rows are numbered correctly.
df1.reset_index(drop = True, inplace = True)
print(df1)
```

```python
#round off years to 1 decimal
df1=df1.round(1)
```
