```python
#dropping colm that are not required
df2=df2.drop(columns =['Indicator Name','Indicator Code'])
print(df2)
```

```python
#Analysed the df & found out that majority countries had more than half data filled hence will substitute NA values in those countries with the row_avg.
#The remaining countries that have NaN values in all/more than half cols will be dropped.
row_avg = df2.mean(axis=1)
print(row_avg)
```

```python
#after analysing df, I found out that countries that have nan val are i.e to be dropped - Andorra[4], American[9],caymen[50],Gibraltor[82],Isle[106],Not classified[108],st.kitts[123],Monaco[147],marshallis[153],northern[162]
#Nauru[177],San Marina[210],Turks[226],Taluvu[243],Brits[253]
#Lets first drop these rows!
df2=df2.drop([df2.index[4],df2.index[9],df2.index[50],df2.index[82],df2.index[106],df2.index[108],df2.index[123],df2.index[147],df2.index[153],df2.index[162],df2.index[177],df2.index[210],df2.index[226],df2.index[243],df2.index[253]])
print(df2)
```

```python
#now I will substitue the rest of the rows(that have more than half colm filled ) i.e having just few NaN values with row_avg using Transpose (T)
df2=df2.T.fillna(df2.mean(axis=1)).T
```

```python
#Now, lets chk the updated isnull values. Should be 0 for all!
df2.isnull().sum()

#Yayy!
```

```python
#Now resetting index so that the rows are numbered correctly.
df2.reset_index(drop = True, inplace = True)
print(df2)
```

```python
#round off years to 1 decimal
df2=df2.round(1)
```
