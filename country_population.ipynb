```python
#dropping extra/unneccessary colm
df=df.drop(columns =['Indicator Name','Indicator Code'])
print(df)
```

```python
#Lets take a look at the null values present before data cleaning
df.isnull().sum()
```

```python
#listing datat types for each col
df.dtypes
```

```python
#Here is a row with all entries as Nan.Will be dropping these rows.
df.loc[108]
```

```python
#droping a specific row
index_to_drop = 108
df=df.drop(108)
print(df)
```

```python
#Analysed the data and found a few rows wherein there are a few Nan entries however, more than half of the entries have values data. Below is 1 such example.
df.loc[193]
```

```python
#I shall take the avg of the rows s.t I may substitute the Nan entries in above example rows (df.loc[193]) with the avg.!
row_avg = df.mean(axis =1)
print(row_avg)
```

```python
#Filling Nan vaues with mean of row using Transpose (T)

#df.fillna(df.mean(axis=1),inplace=True) - Tried this func earlier, however, only a few rows were affected. Hence using the T approach.
df=df.T.fillna(df.mean(axis=1)).T
```

```python
#Now, lets take a look at the null values after cleaning!Should be 0.
df.isnull().sum()
```

```python
#Onelast chk for duplicates
df.duplicated()
duplicate_rows = df.duplicated(subset=['1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','1970','1971','1972','1973','1974','1975','1976','1977','1978','1979','1980','1981','1982','1983','1984','1985','1986','1987','1988','1989','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016'])
print(duplicate_rows)
```

```python
#Resetting index so that the rows are no correctly & dataset is cleaned
df.reset_index(drop = True, inplace = True)
print(df)
```

```python
#round off years to 2 decimal
df=df.round(2)
print(df)
```

```python
years = [str(i) for i in range(1960,2017)]
print(years)
```

```python
melt_df = pd.melt(df,
                  id_vars='Country Code',
                  value_vars=years,
                  var_name='Year',
                  value_name='Population'
                  )
melt_df.head()
```
